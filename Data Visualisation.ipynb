# Visualisation of IRIS data

## Loading required libraries

```python
# Load libraries
import pandas
from pandas.tools.plotting import scatter_matrix
import matplotlib.pyplot as plt
```

## Reading dataset
The dataset can be downloaded from [here](https://archive.ics.uci.edu/ml/datasets/Iris "UCI Website")

```python
names= ['sepal-length', 'sepal-width', 'petal-length', 'petal-width', 'class']
dataset = pandas.read_csv("./Data/iris.data",names=names)
```

## Displaying Statistics

```python
def displayDataStatistics(dataset):
    print("Shape of input data")
    print(dataset.shape)
    print("\nFew entries of dataset\n")
    print(dataset.head(20))
    print("\nBrief description of dataset\n")
    print(dataset.describe())
    print("\nClass statistics\n")
    print(dataset.groupby("class").size())    
displayDataStatistics(dataset)
```

## Visualising  Data

```python
def visualizeData(dataset):
    #multivariate plot
    scatter_matrix(dataset)
    plt.show()
    #histogram
    dataset.hist()
    plt.show()
visualizeData(dataset)
```
